```python
import joblib
from IPython.display import Image, display

models_dir = Path('models')
model_file = models_dir / 'best_model.joblib'
print('Model file:', model_file)

data = joblib.load(model_file)
reports = data.get('reports', {})

for name, r in reports.items():
    print(f"\n---- {name.upper()} ----")
    print(f"Accuracy: {r.get('accuracy')}")
    print(r.get('report'))

# Display confusion matrix images if they exist
for name in ('rf', 'svm', 'knn'):
    img_path = models_dir / f'confusion_{name}.png'
    if img_path.exists():
        print(f"Displaying {img_path.name}:")
        display(Image(filename=str(img_path)))
    else:
        print(f"Image not found: {img_path}")
```
